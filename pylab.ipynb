```python
%matplotlib inline
%config InlineBackend.figure_format='svg'

# Run with "ipython notebook --pylab inline"
import numpy as np
import pylab

x = np.linspace(0, 2 * np.pi, 1000)
pylab.plot(x, np.cos(x), label='cos(x)')
pylab.plot(x, np.sin(x), label='sin(x)')
pylab.axis('tight')
pylab.ylim([-1.1, 1.1])
pylab.legend(loc=0)
pylab.xlabel('x')
pylab.ylabel('y')
pylab.title('trigonometric demo')
```

```python
# scatter plot
title('Pentagram')
x = linspace(0, 1, 1001) * (2 * np.pi)
plot(cos(x), sin(x))
t = linspace(0.0, 1.0, 6) * (4 * np.pi)
plot(sin(t), cos(t))
axis('equal');
```

```python
fig, ax = subplots(figsize=(10,6))
x = linspace(0, 1, 100)
ax.plot(x, x, label=r"$y = \alpha$")
ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.plot(x, x**0.5, label=r"$y = \sqrt{\alpha}$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$', fontsize=18)
ax.set_ylabel(r'$y$', fontsize=18);
ax.set_title(r'Powers of $\alpha$');
```

```python
fig, ax = subplots(figsize=(10,6))
x = logspace(-1, 1, 20)
ax.loglog(x, x, '.-', label=r"$y = \alpha$")
ax.loglog(x, x**2, ':s', label=r"$y = \alpha^2$")
ax.loglog(x, x**3, 'o', label=r"$y = \alpha^3$")
ax.loglog(x, x**0.5, label=r"$y = \sqrt{\alpha}$", color='black')
ax.legend(loc=2) # upper left corner
ax.set_title('log-log graph')
ax.set_xlabel(r'$\alpha$', fontsize=18)
ax.set_ylabel(r'$y$', fontsize=18);
```
